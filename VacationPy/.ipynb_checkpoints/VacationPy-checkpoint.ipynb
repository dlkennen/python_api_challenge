```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#Loading the city weather csv
weather_df = pd.read_csv("cities.csv", index_col = 0)
weather_df.head()
```

```python
#Capitalizing the city names
def capital(x):
    y = str(x).title()
    return y

weather_df["City"] = weather_df.City.apply(capital)

print(weather_df["City"])
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Configure gmaps
gmaps.configure(api_key = g_key)
```

```python
#Use lat and long to create coordinates
locations = weather_df[["Latitude", "Longitude"]]

#Use humidity as weighting
humidV = weather_df["Humidity"].astype(float)

#Configuring the layout of the map
figure_layout = {
    'width': '800px',
    'height': '800px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'}

#Plot heatmap
figV = gmaps.figure(layout=figure_layout, zoom_level = 3, center = (33.7, -84.4))

#Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights = humidV, dissipating = False,
                                max_intensity = 10, point_radius = 2, opacity = .4)

#Add layer
figV.add_layer(heat_layer)

#Display Figure
figV
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
#Creating a subset df based upon maximum temperature, wind speed, cloudiness, and humidity parameter
best_vac_df = weather_df[(weather_df["Temperature (°F)"]<=82) & (weather_df["Temperature (°F)"]>=60)
                        & (weather_df["Wind Speed (mph)"]<=10) 
                        & (weather_df["Cloudiness"] <= 10)
                        & (weather_df["Humidity"] <= 50)]
best_vac_df
```

```python
print(len(best_vac_df))
```

```python
#Fix lat lng coordinates to actually match the center of these cities

params ={"key": g_key}

#Loop through cities and correct coordinates for lat and long
for index, row in best_vac_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    city = row["City"]
    country = row["Country"]
    
    #Add these items to search query
    params["address"] = f"{city}, {country}"
    
    #Make request
    cities_lat_lng = requests.get(base_url, params=params)
    cities_lat_lng = cities_lat_lng.json()
    
    #Overwrite lat lng values with center of the city coordinates
    best_vac_df.loc[index, "Latitude"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]
    best_vac_df.loc[index, "Longitude"] = cities_lat_lng["results"][0]["geometry"]["location"]["lng"]
```

```python
best_vac_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#Re-saving hotel df and inserting a blank column called hotel name
hotel_df = best_vac_df

#Storing NaN in hotel name for ease of removing missing data later
hotel_df["Hotel Name"] = np.NaN

hotel_df
```

```python
#Find the closest hotel
params = {"key": g_key,
          "radius": 5000,
          "types": "lodging"}

#Loop through potential hotel list and search for nearest hotel

for index,row in hotel_df.iterrows():
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    lat = row["Latitude"]
    lng = row["Longitude"]
    
    #Adding city lat and lng coordinates to search query parameters
    params["location"] = f"{lat},{lng}"
    
    #Google Places Request
    hotel_find = requests.get(base_url, params = params)
    hotel_find = hotel_find.json()
    
    # Use a try except to handle missing hotel data
    try:
        hotel_df.loc[index, "Hotel Name"] = hotel_find["results"][0]["name"]
        
    except (KeyError, IndexError):
        print("No data available.")

hotel_df.head()
```

```python
#Drop any cities without a  hotel nearby
hotel_df = hotel_df.dropna()
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]

#I changed the names to match to my hotel_df
locations = hotel_df[["Latitude", "Longitude"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = hotel_info)

# Display figure
figV.add_layer(markers)
figV
```
